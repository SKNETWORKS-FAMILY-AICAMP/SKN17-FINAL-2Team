```python
import re, os, json
```

```python
target_path = "TTA_scaling_v4"
os.makedirs(target_path, exist_ok=True)
```

```python
folder_path = 'TTA_scaling_v3'
folder_list = [f for f in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, f))]
folder_list
```

```python
te = {}
for folder in folder_list:
    file_list = os.listdir(os.path.join(folder_path, folder))
    for file in file_list:
        
        # print(file.replace('_수정', '').replace('.jsonl', '').split('_')[-1])
        # target_file = os.path.join(target_folder, folder, orgin_file_name)
        # print(target_file)
        # print(os.path.join(target_folder, folder, file))
        try:
            te[file.replace('_수정', '').replace('.jsonl', '').split('_')[-1]].append(os.path.join(folder_path, folder, file))
        except:
            te[file.replace('_수정', '').replace('.jsonl', '').split('_')[-1]] = [os.path.join(folder_path, folder, file)]

sorted_dict = {k: te[k] for k in sorted(te.keys())}
sorted_dict
```

```python
for key, values in sorted_dict.items():
    for value in values:
        print(value)
        orgin_file_name = "종합_" + value.split('\\')[-1].replace("drop_", "").replace("keep_", "").replace('_수정', '').split('_')[-1]
        print()
        with open(value, 'r', encoding="utf-8") as f:
            with open(os.path.join(target_path, orgin_file_name), 'a', encoding="utf-8") as f2:
                for line in f:
                    data = json.loads(line)
                    f2.write(json.dumps(data, ensure_ascii=False) + '\n')
```

```python
folder_path = 'TTA_scaling_v4'

file_list = os.listdir(folder_path)
file_list
```

```python
cnt = 0

for file in file_list:
    seen = set()
    duplicates = []
    tete = []
    if "sorted" in file:
        print(os.path.join(folder_path, file))
        # jsonl 파일 열기
        with open(os.path.join(folder_path, file), 'r', encoding='utf-8') as f:
            for line in f:
                try:
                    # 한 줄씩 JSON 객체로 로드
                    data = json.loads(line)
                    
                    # 'content' 값을 기준으로 중복 확인
                    content = data['messages'][0]['content'].replace(" ", "")  # 'content'는 1번 인덱스에 위치
                    
                    if content in seen:
                        duplicates.append(data)
                        cnt += 1
                    else:
                        seen.add(content)
                        tete.append(data)
                except json.JSONDecodeError:
                    continue
        
        with open(os.path.join(folder_path, "_" + file), 'w', encoding="utf-8") as f:
            for te in tete:
                f.write(json.dumps(te, ensure_ascii=False) + '\n')
```
