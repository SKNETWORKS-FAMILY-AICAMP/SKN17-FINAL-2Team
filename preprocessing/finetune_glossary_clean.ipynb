```python
import re, os, json
```

```python
file_name = "finetune_glossary_v2.jsonl"
new_name = "finetune_glossary_scaling.jsonl"
```

```python
with open(new_name, 'w', encoding="utf-8") as new_f:
    with open(file_name, 'r', encoding="utf-8") as f:
        for line in f:
            data = json.loads(line)
            messages = {"messages": [{ "role": "user", "content": data["question"]}, { "role": "assistant", "content": data["answer"] }]}
            new_f.write(json.dumps(messages, ensure_ascii=False) + '\n')
```
